```python
data['playlist_subgenre'].value_counts().plot(kind='bar')
```

```python
# Оставим только строчки с en текстом
data = data[data['language'] == 'en']

#Дропнем столбец с языком
data = data.drop(['language'], axis=1)

data
```

```python
numeric_data = data.select_dtypes(include=[np.number])

# Посмотрим на корреляцию
correlation = numeric_data.corr()

plt.matshow(correlation)
```

```python
# Отсортированные значения кореляции
correlation.unstack().sort_values(ascending=False).drop_duplicates()
```

```python
data.dtypes
```

```python
# Добавим столбец с текстом
data['text'] = data['track_name'] + ' ' + data['track_artist'] + ' ' + data['lyrics']

data['subgenre'] = data['playlist_subgenre']
data['genre'] = data['playlist_genre']

# Удалим ненужные столбцы
data = data.drop(['lyrics', 'track_name', 'track_artist', 'track_id', 'track_album_id', 'track_album_name', 'track_album_release_date', 'playlist_name', 'playlist_id', 'playlist_subgenre', 'playlist_genre'], axis=1)

# Ресетнем индексы
data = data.reset_index(drop=True)
```

```python
data
```

```python
#Понизим регистр текста
data['text'] = data['text'].str.lower()
```

```python
data['genre'].value_counts().plot(kind='bar')
```

```python
# Удалим строки с жанром 'latin'
data = data[data['genre'] != 'latin']

data.value_counts('genre')
```

```python
data['genre'].value_counts().plot(kind='bar')
```

```python
# Сохраним данные
data.to_csv('spotify_songs_cleaned.csv', index=False)
```
